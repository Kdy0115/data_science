pandasと特徴量データの読み込み

```python
# pandasモジュールの読み込み
import pandas as pd
# numpyモジュールの読み込み
import numpy as np
# 村上3次元の特徴量データを読み込み
df = pd.read_excel('Row_Data/特徴量.xlsx',sheet_name=['村上＿3次元','吉井_3次元'])
df_acutual = pd.read_excel('Row_Data/現場データ.xlsx',sheet_name='現場データ')
```

```python
# データ5行までを表示
df = pd.concat([df['村上＿3次元'],df['吉井_3次元'],df_acutual])
df.head()
df = df.reset_index()
```

```python
# 0行目のnanにnameという名前をつける
df['Unnamed: 0'][0] = 'name'
# 特徴量名を確保するためのリスト
column_name = []
# 特徴量名を1つずつ確保
for data in df.loc[0]:
    column_name.append(data)
# 確保した特徴量をカラム名とする
df.columns = column_name
```

```python
df=df[df['服の色']!='服の色']
# インデックスを振りなおす
df = df.reset_index()
# 0行目が削除されたので-1をして調整
df['index'] = df['index']-1
```

```python
df = df.drop(0,axis=1)
df.head()
```

```python
df.columns
```

```python
df.tail()
```

```python
# インデックスを振りなおす
df = df.reset_index()
```

```python
df = df.drop('index',axis=1)
df = df.rename(columns={'level_0':'index'})
```

```python
df.tail()
```

```python
# 特徴量を2次元配列のデータに整形
features_list = []
# 1人分の特徴量データをもつリスト
one_list = []

# Trueから各特長量を大カテゴリごとに数値化してベクトル化する
for column_name, row in df.iterrows():
    one_list.append(row['name'])
    # 名前データは取り除く
    row = row.drop(['index','name'])
    i = 1
    # 特徴量（True,Falseを含む）を1つずつ読み込んでいく
    for key, data in row.items():
        if data==True and key!='服の色' and key!='評価（村上）' and key!='評価（吉井）':
            one_list.append(i)
        i = i+1
        if key == 'ベース形':
            i = 1
            if len(one_list) != 2:
                one_list.append(None)
        elif key == 'ショート':
            i = 1
            if len(one_list) != 3:
                one_list.append(None)
        elif key == '奥二重':
            i = 1
            if len(one_list) != 4:
                one_list.append(None)
        elif key == 'アジア人型':
            i = 1
            if len(one_list) != 5:
                one_list.append(None)
        elif key == 'ニンニク鼻':
            i = 1
            if len(one_list) != 6:
                one_list.append(None)
        elif key == '鷲鼻':
            i = 1
            if len(one_list) != 7:
                one_list.append(None)
        elif key == '低い':
            i = 1
            if len(one_list) != 8:
                one_list.append(None)
        elif key == 'ふつう１':
            i = 1
            if len(one_list) != 9:
                one_list.append(None)
        elif key == '薄い':
            i = 1
            if len(one_list) != 10:
                one_list.append(None)
        elif key == '小さい':
            i = 1
            if len(one_list) != 11:
                one_list.append(None)
        elif key == 'ない':
            i = 1
            if len(one_list) != 12:
                one_list.append(None)
        elif key == '近い':
            i = 1
            if len(one_list) != 13:
                one_list.append(None)
        elif key == 'わからない':
            i = 1
            if len(one_list) != 14:
                one_list.append(None)
        elif key == 'ふつう６':
            i = 1
            if len(one_list) != 15:
                one_list.append(None)
        elif key == 'ない２':
            i = 1
            if len(one_list) != 16:
                one_list.append(None)
        elif key == '遠い２':
            i = 1
            if len(one_list) != 17:
                one_list.append(None)
        elif key == '評価（村上）':
            i = 1
            if data==True:
                one_list.append(1)
            else:
                one_list.append(0)
        elif key == '評価（吉井）':
            i = 1
            if data==True:
                one_list.append(1)
            else:
                one_list.append(0)
        elif key == '服の色':
            if len(one_list)!=20:
                one_list.append(data)
    features_list.append(one_list)
    one_list = []
```

```python
# 整形したデータをDataFrame型へ変換
fea_df = pd.DataFrame(features_list)
```

```python
fea_df = fea_df.dropna(how='all',axis=1)
fea_df = fea_df.drop([21,23,25,27,29,31,33,34,35],axis=1)
```
