# かわいい人の最頻値ベクトルがかわいいと判断したベクトルに含まれるか

```python
# pandasモジュールの読み込み
import pandas as pd
# グラフ描画ライブラリmatplotlibの読み込み
import matplotlib.pyplot as plt
# 数値計算用ライブラリnumpyの読み込み
import numpy as np
# 村上3次元の特徴量データを読み込み
# 引数にindex_col=0を入れなとUnnamed:0が追加されてしまう
# index_colで指定した列をスタートで考えてくれる
df = pd.read_csv('../CSV_DATA/arange_feature_all.csv',index_col=0)
# 欠損値をすべて0に置換
df = df.fillna(0)
# インデックスを振り直す
df = df.reset_index()
```

```python
murakami_pre_value = df[df['Value_of_murakami']==1]
yoshi_pre_value = df[df['Value_of_yoshi']==1]
murakami_pre_value,yoshi_pre_value = murakami_pre_value.drop(['Nose_shape','Nose_len','index','Value_of_murakami','Value_of_yoshi','Cloth_color','name'],axis=1),yoshi_pre_value.drop(['Nose_shape','Nose_len','index','Value_of_murakami','Value_of_yoshi','Cloth_color','name'],axis=1)
murakami_np_array = np.array(murakami_pre_value)
yoshi_np_array = np.array(yoshi_pre_value)
murakami_mode_array = np.array([1.0,1.0,2.0,3.0,1.0,3.0,2.0,2.0,1.0,3.0,1.0,2.0,1.0,1.0])
yoshi_mode_array = np.array([1.0,1.0,2.0,3.0,2.0,3.0,2.0,1.0,1.0,2.0,1.0,1.0,1.0,1.0])

def mode_slm(pre_list,mode_list):
    count = 0
    for array in pre_list:
        if np.allclose(array,mode_list):
            count = count+1
    return count

print('村上の最頻値ベクトルでかわいいと評価された人は{0}人'.format(mode_slm(murakami_np_array,murakami_mode_array)))
print('吉井の最頻値ベクトルでかわいいと評価された人は{0}人'.format(mode_slm(yoshi_np_array,yoshi_mode_array)))
```

```python
murakami_not_value = df[df['Value_of_murakami']==0]
yoshi_not_value = df[df['Value_of_yoshi']==0]
murakami_pre_value,yoshi_pre_value = murakami_not_value.drop(['Nose_shape','Nose_len','index','Value_of_murakami','Value_of_yoshi','Cloth_color','name'],axis=1),yoshi_not_value.drop(['Nose_shape','Nose_len','index','Value_of_murakami','Value_of_yoshi','Cloth_color','name'],axis=1)
murakami_np_array = np.array(murakami_pre_value)
yoshi_np_array = np.array(yoshi_pre_value)
murakami_mode_array = np.array([1.0,1.0,2.0,3.0,1.0,3.0,2.0,2.0,1.0,3.0,1.0,2.0,1.0,1.0])
yoshi_mode_array = np.array([1.0,1.0,2.0,3.0,2.0,3.0,2.0,1.0,1.0,2.0,1.0,1.0,1.0,1.0])

def mode_slm(pre_list,mode_list):
    count = 0
    i = 0
    not_pretty = []
    for array in pre_list:
        if np.allclose(array,mode_list):
            count = count+1
            not_pretty.append(i)
        i = i + 1
    print(not_pretty)
    return count

print('村上の最頻値ベクトルでかわいいと評価された人は{0}人'.format(mode_slm(murakami_np_array,murakami_mode_array)))
print('吉井の最頻値ベクトルでかわいいと評価された人は{0}人'.format(mode_slm(yoshi_np_array,yoshi_mode_array)))
```

```python
murakami_value_data = df[:49]
yoshi_value_data = df[50:98]
rand_value_data = df[100:206]
```

```python
murakami_rand_data = rand_value_data[rand_value_data['Value_of_murakami']==1]
yoshi_rand_data = rand_value_data[rand_value_data['Value_of_yoshi']==1]
```

```python
murakami_rand_data = murakami_rand_data.drop(['index','name','Nose_shape','Nose_len','Value_of_murakami','Value_of_yoshi','Cloth_color'],axis=1)
yoshi_rand_data = yoshi_rand_data.drop(['index','name','Nose_shape','Nose_len','Value_of_murakami','Value_of_yoshi','Cloth_color'],axis=1)
```

```python
mode_rand_murakami_list=[]
for key,data in murakami_rand_data.items():
    mode_list = murakami_rand_data[key].value_counts()
    mode_list = mode_list.to_dict()
    mode_rand_murakami_list.append(max(mode_list,key=mode_list.get))
    
mode_rand_yoshi_list=[]
for key,data in yoshi_rand_data.items():
    mode_list = yoshi_rand_data[key].value_counts()
    mode_list = mode_list.to_dict()
    mode_rand_yoshi_list.append(max(mode_list,key=mode_list.get))
```
