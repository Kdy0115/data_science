```python
m_value,y_value = df[df['Value_of_murakami']=='1'],df[df['Value_of_yoshi'] == 1]

m_value,y_value = m_value.drop(['name','Value_of_murakami','Value_of_yoshi','Cloth_color'],axis=1),y_value.drop(['name','Value_of_murakami','Value_of_yoshi','Cloth_color'],axis=1)
```

## 各特徴量の度数をとって比較してみる

```python
m_fre_dic = {}
y_fre_dic = {}

for column in m_value.columns:
    m_fre_dic[column]=m_value[column].value_counts()
    y_fre_dic[column]=y_value[column].value_counts()
```

```python
print(m_fre_dic['Nose_len'])
print(y_fre_dic['Nose_len'])
```

```python
def plt_show(dic1,dic2,feature):
    y1 = dic1[feature]
    y2 = dic2[feature]

    left = np.arange(len(y1))
    labels = y1.index
    plt.figure(figsize=(10,5),dpi=100)
    plt.bar(left,y1,color='b',width=0.3,label='Murakami_value',align='center')
    plt.bar(left+0.3,y2,color='g',width=0.3,label='Yoshi_value',align='center')

    plt.xticks(left + 0.3/2, labels)
    plt.legend(loc=2)


    plt.show()
```
